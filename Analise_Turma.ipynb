# 06) Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.

# **OBS: Como os dados se referem as notas de uma turma pertencete a uma instituição de ensino, é mais coerente que os NaNs sejam igauis á zero**


```python
dados.isnull().sum() #Quantidade de nulos
new_dados = dados.fillna(0)
new_dados
```

# 7) Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.

```python
index_remocao = new_dados.query('Nome == "Alice" | Nome == "Carlos"').index
new_dados.drop(index_remocao, axis=0, inplace=True)
new_dados
```

# 8)  Aplique um filtro que selecione apenas os alunos que foram aprovados. OBS: Os alunos que tem 7.0 receberam um ponto extra

```python
filtro = new_dados['Aprovado'] == True
alunos_aprovados = new_dados[filtro].replace(7.0, 8.0)
alunos_aprovados
```

# 9) Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=0, sep=';')
```

# 10) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles.  Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

```python
new_dados['ponto_extra'] = new_dados['Notas'].apply(lambda x: x *.40)
new_dados
```

# 11) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.

```python
new_dados['Nota_final'] = new_dados['Notas'] + new_dados['ponto_extra']
new_dados['Nota_final'] = new_dados['Nota_final'].apply(lambda x: 10.0 if(x > 10.0) else x)
new_dados
```

# 12) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:   

#        True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6); False: caso o aluno esteja reprovado (nota final deve ser menor que 6).

```python
new_dados['Aprovado_final'] = new_dados['Nota_final'].apply(lambda x: True if(x >= 6) else False)
new_dados
```

#13) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.


```python
filtro = new_dados['Aprovado_final'] == True
alunos_aprovados = new_dados[filtro]
selecao = alunos_aprovados.query('Aprovado == False')
selecao
```
